```python



def chunk_text_by_token_limit(text, max_tokens=1024):
    """
    Splits a long string by '\r\n', then chunks by token length.
    Removes '\r\n' in final output for clean summarization input.
    """
    comments = [c.strip() for c in text.split("\r\n") if c.strip()]
    chunks = []
    current_chunk = []
    current_token_count = 0

    for comment in comments:
        tokens = tokenizer.encode(comment, add_special_tokens=False)
        if current_token_count + len(tokens) > max_tokens:
            # Start new chunk
            chunks.append(" ".join(current_chunk))
            current_chunk = [comment]
            current_token_count = len(tokens)
        else:
            current_chunk.append(comment)
            current_token_count += len(tokens)

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

```

```python
import pandas as pd
from datasets import Dataset, DatasetDict

# 1) Read the CSV via pandas (this works on any mounted path)
df = pd.read_csv("/content/drive/My Drive/Summarized_PRO_Comments_with_Dimensions.csv")

# 2) Keep only the cols you need
df = df[["Original Comment","Summarized Comment",'Dimensions']]


df['target'] = df['Summarized Comment'] + " | Dimensions: " + df['Dimensions']

# 3) Convert to a Hugging Face Dataset
ds = Dataset.from_pandas(df, preserve_index=False)

# 4) Split into train/validation
splits = ds.train_test_split(test_size=0.1, seed=42)
dataset = DatasetDict({
    "train": splits["train"],
    "validation": splits["test"]
})
```

```python
import pandas as pd
from datasets import Dataset, DatasetDict
from transformers import T5ForConditionalGeneration, T5Tokenizer
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
from transformers import DataCollatorForSeq2Seq
import evaluate

# 6) Load the model and tokenizer
model_name = "google/flan-t5-base"  # Or "facebook/bart-large-cnn" for BART
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# 4) Tokenize function
max_input_length  = 128
max_target_length = 16

def preprocess(ex):
    inp = ex["Original Comment"]
    tgt = ex["target"]  # This combines summary and dimensions
    model_inputs = tokenizer(
        inp,
        max_length=max_input_length,
        truncation=True
    )
    labels = tokenizer(
        tgt,
        max_length=max_target_length,
        truncation=True
    )["input_ids"]

    # Replace pad token id’s in the labels by -100 so they are ignored in the loss
    labels = [(l if l != tokenizer.pad_token_id else -100) for l in labels]
    model_inputs["labels"] = labels
    return model_inputs

tokenized = dataset.map(preprocess, batched=False)

# 8) Data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

# 9) Define metrics (using ROUGE for summarization tasks)
metric = evaluate.load("rouge")

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    labels = [[l for l in seq if l != -100] for seq in labels]
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Compute ROUGE score
    result = metric.compute(predictions=decoded_preds, references=decoded_labels)
    return result

# 10) Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./flant5-hcf-classifier",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    per_device_eval_batch_size=4,
    learning_rate=3e-5,
    num_train_epochs=3,
    logging_steps=100,
    save_total_limit=2,
    fp16=True,               # or False / bf16 on MPS
    report_to=["none"],      # disable W&B
)

# 11) Initialize the trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)


```

```python
# 12) Run the training loop
trainer.train()

# 13) After training, evaluate the model (optional)
#metrics = trainer.evaluate()
#print("Validation metrics:", metrics)

# 14) Save the final fine-tuned model
trainer.save_model("./flant5-hcf-classifier-final")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration


def summarize_with_flan(text, dimension_name, keywords, max_input_tokens=1024):
    # Ensure text is a single string
    if isinstance(text, list):
        text = " ".join(text)

    # Build the prompt
    keyword_str = ", ".join(keywords)
    prefix = (
        f"Summarize the following employee comments with a focus on '{dimension_name}'.\n"
        f"Key aspects: {keyword_str}.\n"
        "Ignore unrelated content.\n\n"
    )
    full_input = prefix + text

    # Tokenize with truncation
    inputs = tokenizer(full_input,
                       max_length=max_input_tokens,
                       truncation=True,
                       return_tensors='pt')

    # Generate
    summary_ids = model.generate(
        inputs.input_ids,
        max_length=200,
        num_beams=4,
        repetition_penalty=1.2,
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
def summarize_all_dimensions(chunks):
    results = {}
    for dim, kws in dimension_keywords.items():
        parts = [summarize_with_flan(c, dim, kws) for c in chunks]
        results[dim] = " ".join(parts)
    return results
```

```python
def summarize_with_flan(text, dimension_name, keywords, max_input_tokens=1024):
    # Ensure text is a single string
    if isinstance(text, list):
        text = " ".join(text)

    # Build the prompt
    keyword_str = ", ".join(keywords)
    prefix = (
        f"Summarize the following employee comments with a focus on '{dimension_name}'.\n"
        f"Key aspects: {keyword_str}.\n"
        "Ignore unrelated content.\n\n"
    )
    full_input = prefix + text

    # Tokenize with truncation
    inputs = tokenizer(full_input,
                       max_length=max_input_tokens,
                       truncation=True,
                       return_tensors='pt')

    # Generate summary
    summary_ids = model.generate(
        inputs.input_ids,
        max_length=200,               # Adjust the length based on your preference
        num_beams=4,                  # Control beam search size
        repetition_penalty=1.2,       # Avoid repetition
        early_stopping=True           # Stop once we get a good summary
    )

    # Decode and return the generated summary
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
def summarize_all_dimensions(chunks):
    results = {}
    # Assuming dimension_keywords is a dictionary with dimension names and their associated keywords
    for dim, kws in dimension_keywords.items():
        # Summarize the comments based on the dimension
        parts = [summarize_with_flan(c, dim, kws) for c in chunks]
        results[dim] = " ".join(parts)  # Combine the results for each dimension
    return results
```
